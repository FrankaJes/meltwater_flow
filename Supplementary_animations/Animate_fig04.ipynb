```python
import numpy as np
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import cmocean as cmo
import matplotlib.animation as animation
from matplotlib.colors import LightSource
from matplotlib import cm
from matplotlib.gridspec import GridSpec

# Import colormaps and limits
import sys
sys.path.append('./')
from colormaps import *
```

```python
data_dir = '../../Zenodo/model_output'

# LA_M, LA_H
M_LAN = xr.open_dataset(f'{data_dir}/LA_M.nc')
H_LAN = xr.open_dataset(f'{data_dir}/LA_H.nc')
M_LAN_calving = xr.open_dataset(f'{data_dir}/calving/LA_M_calving.nc')
H_LAN_calving = xr.open_dataset(f'{data_dir}/calving/LA_H_calving.nc')

la_M = xr.open_dataset(f'{data_dir}/LA_M_laddie_fields.nc')
la_H = xr.open_dataset(f'{data_dir}/LA_H_laddie_fields.nc')
la_M_calving = xr.open_dataset(f'{data_dir}/calving/LA_M_calving_laddie_fields.nc')
la_H_calving = xr.open_dataset(f'{data_dir}/calving/LA_H_calving_laddie_fields.nc')

spinup = xr.open_dataset(f'{data_dir}/spinup.nc')
```

```python
def animate_laddie_melt_velocity_draft_icevel(dsi, dsIMAU, tslice, xslice, fps, savename=None):
    
    """ 
    dsi         :  file LADDIE fields .nc
    dsIMAU      :  file IMAU-ICE output help_fields_ANT.nc
    tslice      :  array of time slices to show, e.g. np.arange(0,1001,10)
    xslice      :  x domain boundaries, e.g. slice(-200000,250000) 
    fps         :  frames per second 
    savename    :  directory and filename to save the animation
    """

    fig,ax = plt.subplots(1,4, figsize=(12,9
                                    ),dpi=200)
    
    snapshots_melt, snapshots_vel, snapshots_draft, snapshots_ice = [],[],[],[]
    quivU, quivV = [],[]
    quivUice, quivVice = [],[]
    mask_gr, mask_oc, mask_GL = [],[],[]
    cont_draft = [[]]
    BMB_vals = []
    im_oc, im_gr = [[],[],[],[]],[[],[],[]]

    for t in tslice:

        # Select time slice and x slice
        ds = dsi.sel(time=t, x = xslice)
        dsIM = dsIMAU.sel(time=t, x = xslice)

        # Transpose data
        newx = ds.y[::-1]
        newy = ds.x
        newmelt = -1*dsIM.BMB.T # * rho_fw / rho_ice ==> melt in m ice per year
        newUt = -ds.Vt.T
        newVt = ds.Ut.T
        newmask = ds.mask.T
        newzb = dsIM.Hs.T-dsIM.Hi_eff_cf.T
        newicevelu = -dsIM['v_surf'].T
        newicevelv = dsIM['u_surf'].T
        newmaskgl = xr.where(dsIM['mask']==7, 1,0).T
        
        if t == 0:
            # Select time slice and x slice
            dsIM = spinup.sel(time=50000, x = xslice)
            newicevelu = -dsIM['v_surf'].T
            newicevelv = dsIM['u_surf'].T
            newmaskgl = xr.where(dsIM['mask']==7, 1,0).T

        snapshots_melt.append(newmelt)
        snapshots_vel.append(np.sqrt(newUt**2+newVt**2))
        snapshots_draft.append(newzb)
        snapshots_ice.append(np.sqrt(newicevelu**2+newicevelv**2))

        BMB_vals.append(np.nansum(newmelt*2000*2000*918*1e-12))

        quiver_slice = (slice(None, None, 3), slice(None, None, 3))  # Slicing every other cell
        quivx = newx[quiver_slice[1]]/1000
        quivy = newy[quiver_slice[0]]/1000

        quivU.append(newUt[quiver_slice])
        quivV.append(newVt[quiver_slice])

        quiver_slice = (slice(None, None, 10), slice(None, None, 10))  # Slicing every other cell
        quivxice = newx[quiver_slice[1]]/1000
        quivyice = newy[quiver_slice[0]]/1000
        
        quivUice.append(newicevelu[quiver_slice])
        quivVice.append(newicevelv[quiver_slice])

        mask_oc.append(xr.where(newmask==0, 1, np.nan))
        mask_gr.append(xr.where(newmask==1, 3, np.nan))
        mask_GL.append(xr.where(newmaskgl==1, 1, np.nan))

    for j, axs in enumerate([ax[0], ax[1], ax[2]]):
        im_oc[j] = axs.pcolormesh(newx/1000, newy/1000,  mask_oc[0],      cmap=cmap_ocean,    norm=norm_ocean, zorder=6)
        im_gr[j] = axs.pcolormesh(newx/1000, newy/1000,  mask_gr[0],      cmap=cmap_ground,   norm=norm_ground, zorder=6)

    # Plot draft
    im_draft = ax[0].pcolormesh( newx/1000, newy/1000, snapshots_draft[0], cmap='cmo.deep_r', norm = mpl.colors.Normalize(vmin=-700, vmax=0))
    cont_draft[0] = ax[0].contour(newx/1000, newy/1000, snapshots_draft[0],  [-300, -250, -225, -200, -175, -150, -125, -100], cmap='inferno_r',zorder=1)
    h, l = cont_draft[0].legend_elements()
    cbar = plt.colorbar(im_draft, ax = ax[0],  orientation='horizontal', label= 'Ice draft depth [m]', pad = 0.05)

    ax[0].set_ylabel('[km]')
    ax[0].set_yticks([-150, -100,-50, 0, 50, 100, 150, 200, 250])
    ax[0].legend(h,['-300 m', '-250 m', '-225 m', '-200 m', '-175 m', '-150 m', '-125 m', '-100 m'],bbox_to_anchor=(-1.0, 1),loc=2)
   
    # Plot melt
    im_melt = ax[1].pcolormesh( newx/1000, newy/1000, snapshots_melt[0], cmap=cmap_BMB, norm = norm_BMB)
    cbar = plt.colorbar(im_melt, ax = ax[1],  ticks=[-10,-1, 0, 1, 10, 100], orientation='horizontal', label= 'Sub-shelf melt [m/yr]', pad = 0.05)
    cbar.ax.set_xticklabels([-10,-1, 0, 1, 10, 100]);
    bmbval = ax[1].text(-38,-195,f'Melt flux: {BMB_vals[0]:.2f} Gt/yr',zorder=7)

    # Plot plume vel
    im_vel = ax[2].pcolormesh( newx/1000, newy/1000, snapshots_vel[0], cmap=cmap_u_lad, norm = norm_u_lad)
    plt.colorbar(im_vel, ax = ax[2], orientation='horizontal', label= 'Layer velocity [m/s]', pad = 0.05)
    Qvel = ax[2].quiver(quivx, quivy, quivU[0], quivV[0], scale=1/1, width=1/100, facecolor='k', zorder=10)

    # Plot ice vel
    im_ice = ax[3].pcolormesh( newx/1000, newy/1000, snapshots_ice[0], cmap=cmap_u_surf, norm = norm_u_surf)
    plt.colorbar(im_ice, ax = ax[3], orientation='horizontal', label= 'Ice surface velocity [m/s]', pad = 0.05)
    Qice = ax[3].quiver(quivxice, quivyice, quivUice[0], quivVice[0], scale=2000, width=1/100, facecolor='white', zorder=10)
    
    im_oc[3] = ax[3].pcolormesh(newx/1000, newy/1000,  mask_oc[0],      cmap=cmap_ocean,    norm=norm_ocean, zorder=9)
    im_gl = ax[3].pcolormesh(newx/1000, newy/1000,  mask_GL[0],      cmap=cmap_GL,   norm=norm_GL, zorder=6)

    plt.suptitle(f'Model year {tslice[0]}', weight='bold', size=15);

    fig.tight_layout()

    
    def animate_func(i):
        if i % fps == 0:
            print( '.', end = '')

        # Clear the previous contours before plotting new ones
        for coll in cont_draft[0].collections:
            coll.remove()
        cont_draft[0] = ax[0].contour(newx/1000, newy/1000, snapshots_draft[i], [-300, -250, -225, -200, -175, -150, -125, -100], cmap='inferno_r',zorder=5)

        for j in range(3):
            im_oc[j].set_array(mask_oc[i])
            im_gr[j].set_array(mask_gr[i])
        im_oc[3].set_array(mask_oc[i])
        im_gl.set_array(mask_GL[i])
        
        im_melt.set_array(snapshots_melt[i]) 
        im_vel.set_array(snapshots_vel[i]) 
        im_draft.set_array(snapshots_draft[i]) 
        im_ice.set_array(snapshots_ice[i]) 

        bmbval.set_text(f'Melt flux: {BMB_vals[i]:.2f} Gt/yr')

        Qvel.set_UVC(quivU[i],quivV[i])
        Qice.set_UVC(quivUice[i],quivVice[i])

        fig.suptitle(f'Model year {tslice[i]}', weight='bold', size=20);

        return cont_draft[0].collections


    anim = animation.FuncAnimation(
                               fig, 
                               animate_func, 
                               frames = len(tslice*10),
                               interval = 1000 / 400, blit=False# in ms
                               )

    
    anim.save(f'{savename}.mp4', fps=fps, extra_args=['-vcodec', 'libx264']) 

    return 
```

```python
# Create animations LA_M and LA_H
animate_laddie_melt_velocity_draft_icevel(la_M, M_LAN, np.arange(0,1001,10), slice(-200000,250000), fps=5, savename='LA_M_animation_fields')
animate_laddie_melt_velocity_draft_icevel(la_H, H_LAN, np.arange(0,1001,10), slice(-200000,250000), fps=5, savename='LA_H_animation_fields')
```

```python
# Create animations LA_M_calving and LA_H_calving
animate_laddie_melt_velocity_draft_icevel(la_M_calving, M_LAN_calving, np.arange(0,1001,10), slice(-200000,250000), fps=5, savename='LA_M_calving_animation_fields')
animate_laddie_melt_velocity_draft_icevel(la_H_calving, H_LAN_calving, np.arange(0,1001,10), slice(-200000,250000), fps=5, savename='LA_H_calving_animation_fields')
```
